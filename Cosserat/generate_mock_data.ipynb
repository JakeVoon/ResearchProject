# Generating Mock Data
This notebook generates mock data for use in the project talk

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Accuracy

```python
grating = np.linspace(0,7,100)

force = [0]*25 + list(np.linspace(0,2,25)) + list(np.linspace(2,0,25)) + [0]*25
mock = [x-0.075 for x in force]

plt.plot(grating,force,label="Ground Truth");
plt.plot(grating,mock,label="Mock Measurement");
plt.title('Mock Data');
plt.legend();
plt.xlabel("Seconds (s)")
plt.ylabel("Force (N)")
```

## Reaction Time

```python
time = np.linspace(0,30,100)

force = [0]*25 + list(np.linspace(0,2,50)) + [2]*25
mock = force

plt.plot(time,force,label="Ground Truth");
plt.plot(time+1,mock,label="Mock Measurement");
plt.title('Mock Data');
plt.legend();
plt.xlabel("Seconds (s)")
plt.ylabel("Force (N)")
```
